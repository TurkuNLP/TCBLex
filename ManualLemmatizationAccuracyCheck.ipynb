```python
#Imports
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
from scripts import bookdatafunctions as bdf
from TCBC_tools import Structure
from pprint import pprint
from collections import Counter
import random
```

```python
corpus = Structure.initBooksFromConllus("Conllus_v-1-0")
```

```python
id_trees = {x:Structure.buildIdTreeFromConllu(corpus[x]) for x in corpus}
```

```python
all_sentence_heads = [(x,id_trees[x][y]) for x in id_trees for y in id_trees[x]]
```

```python
#Shuffle
random.shuffle(all_sentence_heads)
#Sample sentences
our_sample = random.sample(all_sentence_heads, k=300)
```

```python
all_sample_lines = [y for x in our_sample for y in x[1]]
```

```python
test_conllu = ""
#Building the test-set Conllu from our sample
for id,tree in our_sample:
    df = corpus[id]
    lines = list(tree.keys())
    test_conllu += "#This test set contains a total of "+str(len(all_sample_lines))+" lines \n"
    test_conllu += "#book_id="+str(id)+" lines:"
    for line in lines:
        test_conllu += " "+str(line)
    test_conllu += "\n"
    for line_num in lines:
        for item in df.iloc[line_num]:
            test_conllu += str(item)+"\t"
        test_conllu += "\n"
```

```python
incorrect_counter = 0
non_puncts = 0
for x in range(len(lines_og)):
    if lines_og[x][1] != "PUNCT" and lines_og[x][1] != "SYM":
        non_puncts += 1
    if lines_og[x][0] != lines_corr[x][0]:
        incorrect_counter += 1

print("Incorrect lemmas: ",incorrect_counter)
print("Accurate lemmas percentage: ",(len(lines_og)-incorrect_counter)/len(lines_og))
print("Accurate lemmas percentage when excluding PUNCT/SYM: ",(non_puncts-incorrect_counter)/non_puncts)
```
